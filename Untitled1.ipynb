```python
def text_clean(synopsis,stopWords,drop_punc=True,drop_num=True,stem=True,stopper=True):
        token_pattern = re.compile('(?u)\\b\\w\\w+\\b')
        if drop_punc:
            punc_translator = str.maketrans(string.punctuation, ' '*len(string.punctuation))
            synopsis=synopsis.translate(punc_translator)
        if drop_num:
        # remove standalone numbers
            num_translator = str.maketrans('', '', string.digits)
            synopsis=synopsis.translate(num_translator)
        if stem:
            stemmer=PorterStemmer()
            synopsis=' '.join([stemmer.stem(term) for term in list(synopsis.split())])
        if stopper:
            synopsis = ' '.join([term for term in token_pattern.findall(synopsis.lower()) if term not in stopWords])
        return synopsis 
    
```

```python
#remove stopwords, numbers, punctuation and stem
stopWords = set(stopwords.words('english'))
stopWords.update(['set'],['based'],['mins'],['company'],[string.ascii_letters])

synopsis_old=synopsis.copy()
synopsis=[]
#Remvoe named entities
for sent in synopsis_old:
    ed =[word for word,tag in pos_tag(sent.split()) if tag != 'NNP' and tag != 'NNPS']
    synopsis.append(' '.join(ed))

clean_synopsis=[]
for i in range(len(synopsis)):
    clean_synopsis.append(text_clean(synopsis[i],drop_punc=True,drop_num=True,stem=True,stopper=True,stopWords=stopWords))

# synopsis_df = pd.DataFrame({'title_':data.title,'synopsis':clean_synopsis,'release_yr':data.releaseYear})
# synopsis_df['title'] = synopsis_df.title_ + " " + synopsis_df.release_yr.astype(str)
# synopsis_df.drop('title_',axis=1,inplace=True)
# synopsis_df['prog_uuid'] = data.programmeUuid
# synopsis_df['series_uuid']=data.seriesUuid
# # A few movies with the same name and release year
# full_df= pd.merge(left=synopsis_df,right=ind_tags,on=['prog_uuid','prog_uuid'],how='right')
 
```

```python
import gensim
from gensim.models.doc2vec import TaggedDocument
conc_syn = pd.read_pickle("/Users/jbo50/Documents/AutoTagging/research-AutoTagging/data/concatenated_synopsis.pkl")
```

```python
taggeddoc=[]
for index,syn in enumerate(conc_syn.synopsis):
    td = TaggedDocument(syn.split(),[conc_syn.tags[index]])
    # for later versions, you may want to use: td = TaggedDocument(gensim.utils.to_unicode(str.encode(' '.join(stemmed_tokens))).split(),[str(index)])
    taggeddoc.append(td)
 
```

```python
# build the model
model = gensim.models.Doc2Vec(dm = 0, alpha=0.025, size= 50, min_alpha=0.025, min_count=0)
model.build_vocab(taggeddoc)
```

```python
#training of model
for epoch in range(100):
 print('iteration'+str(epoch+1))
 model.train(taggeddoc, total_examples=model.corpus_count,epochs=model.iter)
 model.alpha -= 0.002
 model.min_alpha = model.alpha
```
